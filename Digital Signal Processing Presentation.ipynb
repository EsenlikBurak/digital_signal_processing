```python
import scipy.io.wavfile 
import scipy.signal
import numpy as np
import matplotlib.pyplot as plt


#Bu bir ecg verisi
dosya_yolu = r"C:\Users\ASUS\Downloads\ecg.wav"

sampleRate, data = scipy.io.wavfile.read(dosya_yolu)
#WAV dosyasn okumak iin scipy.io.wavfile.read() ilevini kullanyoruz. Dosyann yolunu belirttikten sonra rnekleme frekans
#sampleRate ve sinyal verisini data eklinde dndryoruz
times = np.arange(len(data))/sampleRate
#sinyalin sresini times diye isimlendirerek hesaplamak iin arange Len data bl rnekleme frekans yaptk
#Orijinal sinyali grselletirdim
plt.figure(figsize=(12,4))
plt.plot(times, data)
plt.title("orijinal ecg sinyali")
plt.xlabel("zaman")
plt.ylabel("genlik")
```

```python
def frekans_cevabi_cizdir (x_n, fs, frekans_nokta_sayisi=500):
 X_f = np.fft.fft(x_n, frekans_nokta_sayisi*2)
 X_f = X_f[:frekans_nokta_sayisi]
 f_ax = np.linspace(0, fs//2, frekans_nokta_sayisi)

 genlik_w= np.abs (X_f)
 faz_w= np.unwrap(np.angle(X_f))

 plt.figure(figsize=(12,3))
    
 plt.subplot(1,2,1)
 plt.title('Genlik spektrumu')
 plt.plot(f_ax,genlik_w, 'b')
 plt.ylabel('|X(f) |')
 plt.xlabel('frekans [Hz]')
 plt.grid()
           
 plt.subplot(1,2,2)
 plt.title('Faz spektrumu')
 plt.plot(f_ax, faz_w, 'g')
 plt.ylabel('faz(X(f)) (radyan)') 
 plt.xlabel('frekans [Hz]')
 plt.grid()
 plt.show()
```

```python
frekans_cevabi_cizdir(data,sampleRate)
```

```python
data_mean_corrected = data-np.mean(data)
```

```python
frekans_cevabi_cizdir(data_mean_corrected,sampleRate)
```

```python
#g hatlarndan kaynaklanan 50 hz bileeninin temizlenmesini bandstop oygulayarak bu kodda yaptk. 
f_c = np.array([49,51]) #cut off 49-51 aral
f_max= sampleRate / 2 # max frekans rnekleme frekans bl 2
f_norm=f_c / f_max #normalize frekans ise cutt of bl maximum frekans
b, a = scipy.signal.butter (3, f_norm, btype='bandstop') #bandstopu uyguladk
filtered = scipy.signal.filtfilt(b, a, data_mean_corrected )
```

```python
frekans_cevabi_cizdir(filtered,sampleRate)


# ## Low pass, high pass, band pass filter

# Bu ksmda dk geiren, yksek geiren ve band geiren filtreleri tasarlayp Ekg sinyaline uyguladk 
```

```python
b, a = scipy.signal.butter(3, 0.05, 'lowpass')#25 hzin altndaki sinyaller geirilip grntlendi (max frekansn yzde 5i)
filteredLowPass = scipy.signal.filtfilt(b, a, data)

b, a = scipy.signal.butter(3, 0.05, 'highpass')#25 hzin stndeki sinyaller geirilip grntlendi
filteredHighPass = scipy.signal.filtfilt(b, a, data)

b, a = scipy.signal.butter(3, [.01, .05], 'band')#5-25 hz aralndaki sinyalleri geiren bandpass filtre
filteredBandPass = scipy.signal.lfilter(b, a, data)
# filtreleri uyguladktan sonra plot aamasna getik
```

```python
# orijinal sinyal ve filtrelenmi tm sinyalleri ayr ayr gsterdik.
plt.figure(figsize=(12, 6))

plt.subplot(221)
plt.plot(times, data)
plt.title("Original ECG Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(222)
plt.plot(times, filteredLowPass)
plt.title("Lowpass Filtered ECG Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(223)
plt.plot(times, filteredHighPass)
plt.title("Highpass Filtered ECG Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.subplot(224)
plt.plot(times, filteredBandPass)
plt.title("Bandpass Filtered ECG Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()


# lowpass filtera baktmzda sinyaln max frekansnn yzde 5inden dk frekanstaki bileeenlerini grntledik. yksek frekansl grltlerin azaldn ve daha temiz bi sinyal olutuunu grdk

# highpass filtera baktmzda sinyaln max frekansnn yzde 5inden yksek frekanstaki bileeenlerini grntledik sinyaldeki dk frekanslarn azaldn ve daha yksek frekansl bir sinyal olutuunu grdk

# bandpass filtermzda da 5-25 Hz aralndaki frekanslar geiren bir bandpass filtre oluturup grntledik.
```

```python
frekans_cevabi_cizdir(filteredLowPass,sampleRate) #frekans cevab izdirilince 25 hz stn filtreledi
```

```python
frekans_cevabi_cizdir(filteredHighPass,sampleRate) #frekans cevab izdirilince 25 hz altn filtreledi
```

```python
frekans_cevabi_cizdir(filteredBandPass,sampleRate) #frekans cevab izdirilince 5-25 hz altn filtreledi
```

```python
# FIR filtre tasarm iin gerekli ktphane
from scipy.signal import firwin

# FIR filtre tasarm parametreleri
bant_genisligi = 0.1  # Bant genilii (0-1 arasnda bir deer)

# FIR filtre katsaylarn tasarla
filtre_sira = 101  # FIR filtre sras (tek say olmal)
fir_katsayilari = firwin(filtre_sira, bant_genisligi)
```

```python
# Filtreleme ilemi
filtrelenmis_sinyal = np.convolve(data, fir_katsayilari, mode='same')

# Filtrelenmi sinyali grselletirme
plt.figure(figsize=(12, 4))
plt.plot(times, filtrelenmis_sinyal)
plt.title("Filtrelenmi EKG Sinyali (Dk Geiren FIR Filtre)")
plt.xlabel("Zaman")
plt.ylabel("Genlik")
plt.show()
```

```python
# Orijinal sinyali grselletirme
plt.figure(figsize=(12, 4))
plt.plot(times, data, label="Orijinal Sinyal", color="blue")

# Filtrelenmi sinyali grselletirme
plt.plot(times, filtrelenmis_sinyal, label="Filtrelenmi Sinyal", color="red")

plt.title("Filtrelenmi EKG Sinyali ve Orijinal Sinyal")
plt.xlabel("Zaman")
plt.ylabel("Genlik")
plt.legend()
plt.show()
```

```python
from scipy.signal import find_peaks

# Tepe noktalarn bulma
peaks, _ = find_peaks(filtrelenmis_sinyal, distance=700)

# Orijinal sinyali grselletirme
plt.figure(figsize=(12, 4))
plt.plot(times, filtrelenmis_sinyal)
plt.title("R Tepe Deerleri)")
plt.xlabel("Zaman")
plt.ylabel("Genlik")

# Tepe noktalarn iaretlemek iin scatter plot kullanma
plt.scatter(times[peaks], filtrelenmis_sinyal[peaks], color='red', label='Tepe Noktalar')

plt.legend()
plt.show()
```

```python
# Genlik spektrumunu hesaplama
genlik_spektrumu = np.abs(np.fft.fft(data))
genlik_spektrumu = genlik_spektrumu[:len(genlik_spektrumu)//2]  # Yalnzca pozitif frekanslar

# Genlik spektrumunu normalize etme
genlik_spektrumu = genlik_spektrumu / np.max(genlik_spektrumu)

# Faz spektrumunu hesaplama
faz_spektrumu = np.angle(np.fft.fft(data))
faz_spektrumu = faz_spektrumu[:len(faz_spektrumu)//2]  # Yalnzca pozitif frekanslar

# Faz spektrumunu normalize etme
faz_spektrumu = (faz_spektrumu + np.pi) / (2 * np.pi)

# Genlik spektrumunu izdirme
frekanslar = np.fft.fftfreq(len(data), d=1/sampleRate)
frekanslar = frekanslar[:len(frekanslar)//2]  # Yalnzca pozitif frekanslar

plt.figure(figsize=(12, 4))
plt.plot(frekanslar, genlik_spektrumu)
plt.title("Genlik Spektrumu")
plt.xlabel("Frekans (Hz)")
plt.ylabel("Normalized Genlik")
plt.show()

# Faz spektrumunu izdirme
plt.figure(figsize=(12, 4))
plt.plot(frekanslar, faz_spektrumu)
plt.title("Faz Spektrumu")
plt.xlabel("Frekans (Hz)")
plt.ylabel("Normalized Faz")
plt.show()
```

```python
# Orijinal sinyalin genlik spektrumu
orijinal_genlik_spektrumu = np.abs(np.fft.fft(data))
orijinal_genlik_spektrumu = orijinal_genlik_spektrumu[:len(orijinal_genlik_spektrumu)//2]  # Yalnzca pozitif frekanslar

# Filtrelenmi sinyalin genlik spektrumu
filtrelenmis_genlik_spektrumu = np.abs(np.fft.fft(filtrelenmis_sinyal))
filtrelenmis_genlik_spektrumu = filtrelenmis_genlik_spektrumu[:len(filtrelenmis_genlik_spektrumu)//2]  # Yalnzca pozitif frekanslar

# Frekanslar
frekanslar = np.fft.fftfreq(len(data), d=1/sampleRate)
frekanslar = frekanslar[:len(frekanslar)//2]  # Yalnzca pozitif frekanslar

# Genlik spektrumlarn st ste izdirme
plt.figure(figsize=(12, 4))
plt.plot(frekanslar, orijinal_genlik_spektrumu, label="Orijinal Sinyal")
plt.plot(frekanslar, filtrelenmis_genlik_spektrumu, label="Filtrelenmi Sinyal")
plt.title("Orijinal ve Filtrelenmi Sinyallerin Genlik Spektrumu")
plt.xlabel("Frekans (Hz)")
plt.ylabel("Genlik")
plt.legend()
plt.show()
```

```python
#IIR Filtre tasarm
from scipy.signal import butter, filtfilt

# Filtre tasarm iin parametreler
cutoff_freq = 50  # Kesim frekans (Hz)
filter_order = 4  # Filtre sras

# IIR filtresini tasarlama
b, a = butter(filter_order, cutoff_freq, fs=sampleRate, btype='low', analog=False, output='ba')

# Filtrelenmi sinyali elde etme
iir_filtrelenmis_sinyal = filtfilt(b, a, data)

# Orijinal ve filtrelenmi sinyalleri izdirme
plt.figure(figsize=(12, 4))
plt.plot(times, data, label='Orijinal Sinyal')
plt.plot(times, iir_filtrelenmis_sinyal, label='Filtrelenmi Sinyal')
plt.title('IIR Filtresi ile Dk Geiren Filtre Uygulamas')
plt.xlabel('Zaman (s)')
plt.ylabel('Genlik')
plt.legend()
plt.show()
```
